```python
from sklearn.datasets import load_boston
from sklearn.preprocessing import PolynomialFeatures

import pandas as pd
```

```python
boston_dataset=load_boston()
```

```python
boston_dataset.data
```

```python
boston_dataset.data.shape
```

```python
boston_dataset.feature_names
```

```python
Polynomial_transformer=PolynomialFeatures(2) # 다항데이터로 변환 , 2차함수이므로 2로 기입
```

```python
Polynomial_data=Polynomial_transformer.fit_transform(boston_dataset.data)
```

```python
Polynomial_data
```

```python
Polynomial_data.shape
```

```python
Polynomial_feature_names=Polynomial_transformer.get_feature_names(boston_dataset.feature_names)  # 속성 이름들을 받아오겠다. 
```

```python
Polynomial_feature_names
```

```python
X=pd.DataFrame(Polynomial_data, columns=Polynomial_feature_names)
```
