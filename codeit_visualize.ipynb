# 예제

```python
%matplotlib inline
import pandas as pd

df = pd.read_csv('data/silicon_valley_details.csv')

# 코드를 작성하세요.
df['company'].value_counts()
adobe=df['company']=='Adobe'
total=df['race']=='Overall_totals'
df_ad_total=df[adobe & total]
df_ad_total2=(df_ad_total['job_category']!='Totals') & (df_ad_total['job_category']!='Previous_totals') 
df_ad_total3=df_ad_total[df_ad_total2]
total4=df_ad_total3['count']!=0
total5=df_ad_total3[total4]
total6=total5[['job_category','count']]
total6.set_index('job_category',inplace=True)
total6.plot(kind='pie',y='count')
```
